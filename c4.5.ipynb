```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import os
import math
import sys
```

```python
class c4_5:
    def __init__(self,file,configDICT):
        self.loadData(file,configDICT)
        print("init")
        
    def loadData(self,file,configDICT):
        self.data = pd.read_csv(file)
        self.config = configDICT
        self.attrs = configDICT["attrs"]
        self.classes = configDICT["classes"]
        self.cls_attr = configDICT["cls_attr"]
    
    def information(self,attr):
        info = {}
        attrEntropy = 0
        total = 0
        for i,row in self.data.iterrows():
            total = total + 1
            if row[attr] in self.attrs[attr]["labels"]:
                if row[attr] in info:
                    info[row[attr]]["classes"][row[self.cls_attr]] = info[row[attr]]["classes"][row[self.cls_attr]] + 1
                    info[row[attr]]["total"] = info[row[attr]]["total"] + 1
                else:
                    info[row[attr]] = {"total": 1,"entropy":0,"classes":{}}
                    for cls in self.classes:
                        info[row[attr]]["classes"][cls] = 1
        for key in info:
            for cls in info[key]["classes"]:
                info[key]["entropy"] = (info[key]["entropy"] 
                                        + (-(info[key]["classes"][cls]/info[key]["total"]) * 
                                               math.log2(info[key]["classes"][cls]/info[key]["total"])))
            attrEntropy = attrEntropy + (info[key]["total"]/total)*info[key]["entropy"]
            
        
        return info,attrEntropy
    
    def getInfoData(self):
        cls = {}
        entropy = 0
        total = 0
        for i,row in self.data.iterrows():
            total = total + 1
            if row[self.cls_attr] in cls:
                cls[row[self.cls_attr]] = cls[row[self.cls_attr]] + 1
            else:
                cls[row[self.cls_attr]] = 1
                
        for key in cls:
            entropy = entropy + (-(cls[key]/total)*math.log2(cls[key]/total))
        return entropy
    
    def fit(self):
        stack = [{"depth":0,"selected_attr":{},"selected_label":{}}]
        depth = 0
        while len(stack) > 0 :
            item = stack[-1]
            print("the depth {}:".format(item["depth"]))
            globalEntropy = self.getInfoData()
            print(globalEntropy)
            minEntropy = sys.maxsize
            minArgEntropy = None
            for kattr in self.attrs:
                if kattr not in item["selected_attr"]:
                    item["selected_attr"][kattr]= depth
                    if self.attrs[kattr]["type"] == "categorical":
                        #labels = self.attrs[kattr]["labels"]
                        info,attrEntropy = self.information(kattr)
                        if minEntropy > globalEntropy - attrEntropy:
                            minArgEntropy = [kattr,info]
                            minEntropy = globalEntropy - attrEntropy
            print(minEntropy)
            return
                        #for klabel in labels:
                            #if klabel not in item["selected_label"]:
                                
                    
        #por cada atributo que no se encuentre seleccionado
            # obtener etiquetas del atributo
            # seleccionamos la etiqueta mejor
            # 
```
